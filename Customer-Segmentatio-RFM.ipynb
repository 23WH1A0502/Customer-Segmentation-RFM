```python
import pandas as pd

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

reference_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
reference_date
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,  
    'InvoiceNo': 'nunique',                                    
    'Revenue': 'sum'                                        
})
rfm
```

```python
rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.head()
```

```python
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1,2,3,4])
rfm['M_score'] = pd.qcut(rfm['Monetary'].rank(method='first'), 4, labels=[1,2,3,4])
```

```python
rfm['Frequency'].value_counts().sort_index()
```

```python
rfm['Frequency'].value_counts().sort_index()
```

```python
# For Frequency
rfm['F_score'] = pd.cut(
    rfm['Frequency'],
    bins=[0,1,2,5, rfm['Frequency'].max()],
    labels=[1,2,3,4],
    include_lowest=True
)

# For Monetary
rfm['M_score'] = pd.cut(
    rfm['Monetary'],
    bins=[0,100,500,1000, rfm['Monetary'].max()],
    labels=[1,2,3,4],
    include_lowest=True
)
```

```python
rfm['R_score'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])
```

```python
rfm['RFM_Score'] = (
    rfm['R_score'].astype(str) +
    rfm['F_score'].astype(str) +
    rfm['M_score'].astype(str)
)
```

```python
rfm
```

```python
def segment_customer(row):
    if row['RFM_Score'] >= '444':
        return 'Best Customers'
    elif row['F_score'] == '4':
        return 'Loyal Customers'
    elif row['R_score'] == '1':
        return 'At Risk'
    else:
        return 'Average Customers'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)
rfm
```

```python
rfm['Segment'].value_counts()
```

```python
import matplotlib.pyplot as plt

rfm['Segment'].value_counts().plot(kind='bar')
plt.title('Customer Segmentation (RFM)')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.show()
```

```python
rfm.groupby('Segment')['Monetary'].sum().sort_values(ascending=False)
```

```python
rfm.groupby('Segment')['Monetary'].sum().plot(kind='bar')
plt.title('Revenue Contribution by Customer Segment')
plt.ylabel('Total Revenue')
plt.show()
```

🔹 Business Insights

1.Best Customers contribute the highest revenue, even though they are fewer in number.
This shows that retaining these customers is very important for business growth.

2.Loyal Customers make frequent purchases, indicating strong engagement with the business.
They can be targeted with reward programs or special offers to increase retention.

3.At-Risk and Low-Value Customers contribute less revenue, but they form a large group.
Businesses can use discounts or reminder campaigns to re-engage them.
